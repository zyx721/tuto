```python
import pandas as pd
```

```python
df = pd.read_csv("carprices.csv")
```

```python
df
```

```python
dm = pd.get_dummies(df['Car Model'])
```

```python
dm
```

```python
dm = dm.astype(int)
```

```python
dm
```

```python
mg = pd.concat([df,dm],axis='columns')
```

mg

```python
mg
```

```python
mg = mg.drop(['Car Model'],axis='columns')
```

```python
mg = mg.drop(['Mercedez Benz C class'],axis='columns')
```

mg

```python
mg
```

```python
x = mg.drop(['Sell Price($)'],axis='columns')
```

```python
y =mg['Sell Price($)']
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(x,y)
```

```python
model.predict(x)
```

```python
model.score(x,y)
```

```python
import joblib
```

```python
joblib.dump(model, 'model_joblib')
```
