```python
import pandas as pd
```

```python
from sklearn.datasets import load_digits
```

```python
ld = load_digits()
```

```python
from sklearn.model_selection import cross_val_score
```

```python
import matplotlib.pyplot as plt
```

```python
plt.gray()
plt.matshow(ld.images[0])
plt.matshow(ld.images[10])
plt.matshow(ld.images[20])
```

```python
df = pd.DataFrame(ld.data)
```

```python
df
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
cross_val_score(RandomForestClassifier(n_estimators=500),ld.data,ld.target,cv=10)
```

```python
from sklearn.model_selection import train_test_split
```

```python
tr1,ts1,tr2,ts2= train_test_split(ld.data,ld.target,test_size=0.2)
```

```python
model = RandomForestClassifier()
```

```python
model.fit(tr1,tr2)
```

```python
model.score(ts1,ts2)
```

```python
yp = model.predict(ts1)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(ts2,yp)
```

```python
import seaborn as sns
```

```python
sns.heatmap(cm,annot=True)
```
