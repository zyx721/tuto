```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
df = pd.read_csv("hiring.csv")
```

```python
ndf = df.dropna()
```

```python
word_to_int = {
    'one': 1,
    'two': 2,
    'three': 3,
    'four': 4,
    'five': 5,
    'six':6,
    'seven':7,
    'ten':10,
    'eleven':11,
}
```

```python
df['experience'] = df['experience'].apply(lambda x: word_to_int.get(x, x))
```

```python
df
```

```python
ndf = df.dropna()
```

```python
ndf
```

```python
firstreg = linear_model.LinearRegression()
```

```python
firstreg.fit(ndf[['test_score(out of 10)','interview_score(out of 10)','salary($)']],ndf.experience)
```

```python
firstreg.predict([[8,6,85000]])
```
